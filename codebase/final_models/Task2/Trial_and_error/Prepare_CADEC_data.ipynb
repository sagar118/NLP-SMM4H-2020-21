```python
import pandas
import zipfile
import spacy
import warnings
import pandas as pd
import numpy as np

from spacy.training import offsets_to_biluo_tags
from glob import glob

warnings.filterwarnings("ignore")
pd.options.display.max_rows = None
```

```python
with zipfile.ZipFile('CADEC.v2.zip', 'r') as zip_ref:
    zip_ref.extractall('.')
```

```python
# dir_path = './CADEC.v1/MedDRA/'
dir_path = './cadec/meddra/'
text_dir = './cadec/text/'
```

```python
filepath_list = list()
text_pathlist = list()

for filepath in glob(dir_path + "*.ann"):
    # print(filepath)
    filepath_list.append(filepath)

for filepath in glob(text_dir + "*.txt"):
    # print(filepath)
    text_pathlist.append(filepath)
```

```python
extractions_details = list()
text_details = list()
```

```python
i = 0
for filepath in filepath_list:
    with open(filepath, 'r') as handle:
        lines = handle.readlines()
    
    extraction = list()
    for index, line in enumerate(lines):
        if i == 49:
            print(line)
        if line.count("\t") == 2:
            extraction.append(line.strip("\n").split("\t"))
            i += 1
        else:
            continue

    # extraction = [line.strip("\n").split("\t") if '\t' in line else continue for line in lines]
    for ext in extraction:
        # print(ext, filepath.split('/')[-1])
        if ';' in ext[1]:
            ext.extend(ext[1].split(';'))
            temp1, temp2, temp3 = ext[3].rsplit(' ',2)
            del ext[3]
            ext.insert(3, temp2+' '+temp3)
        elif len(ext[1].split()) == 3:
            ext.extend(ext[1].split(' ', 1)[1:])
        elif '+' in ext[1]:
            ext.extend(ext[1].split(' ', 3)[3:])
        else:
            ext.extend(ext[1].split(' ', 1)[1:])
        del ext[1]
    
    filename = filepath.split('/')[-1]
    with open(text_dir + filename.strip(".ann") + '.txt', 'r') as handle:
        text = handle.readlines()
    
    text = " ".join([row.strip("\n") for row in text])
    
    for ext in extraction:
        extractions_details.append(ext)
        text_details.append(text)
```

```python
cadec_data = pd.DataFrame(columns=["text", "extraction", "indexes"])
```

```python
cadec_data.text = text_details
cadec_data.extraction = [row[1] for row in extractions_details]
cadec_data.indexes = [row[2:] for row in extractions_details]
```

```python
cadec_data
```

```python
nlp = spacy.load('en_core_web_trf')
ner = nlp.get_pipe('ner')
```

```python
TRAIN_DATA = []
text_tokens = list()
tags_list = list()
```

```python
for index, row in enumerate(cadec_data.itertuples()):
    TRAIN_DATA.append((row[1],{
        'entities': [(int(value.split()[0]), int(value.split()[1]), 'ADR') for value in row[3]]
    }))
```

```python
for text, annot in TRAIN_DATA:
    doc = nlp(text)
    text_tokens.append([token for token in doc])
    tags = offsets_to_biluo_tags(doc, annot['entities'])
    bio_tags = list(map(lambda tag: tag.replace("L", "I").replace("U", "B"), tags))
    bio_tags = ['O' if tag == '-' else tag for tag in bio_tags]
    tags_list.append(bio_tags)
```

```python
cadec_data['text_tokens'] = text_tokens
cadec_data['bio_tags'] = tags_list
```

```python
cadec_data.to_excel("cadec_data.xlsx")
```
